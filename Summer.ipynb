# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np

df = pd.read_csv("summer.csv")
```


1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = df.values

lst = []


for sport in df['Sport'].unique():
    count = 0
    for i in range(len(data)):
        if(data[i][2] == sport and data[i][8] == 'Gold'):
                count = count + 1
    lst.append([sport,count])

    
df1 = pd.DataFrame(lst, columns = ['Sport','Gold medals']).sort_values(by='Gold medals',ascending=False).head()

print('Top 5 sports having most number of Gold medals:',end='\n\n')
print(df1,end='\n\n')
    
df1.plot(x='Sport', y='Gold medals',kind='bar',figsize=(10,5))
```

7. Which event is most popular in terms on number of players? (Top 5)

```python
data = df.values

lst = []


for event in df['Event'].unique():
    count = 0
    for i in range(len(data)):
        if(data[i][7] == event):
                count = count + 1
    lst.append([event,count])
    
   

df5 = pd.DataFrame(lst, columns = ['Event','Players']).sort_values(by='Players',ascending=False).head()

print('Top 5 Events which are most popular in terms of number of Players:',end='\n\n')
print(df5,end='\n\n')

df5.plot(x='Event', y='Players',kind='bar',figsize=(10,5))
```
